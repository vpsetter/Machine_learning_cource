```python
import pandas as pd
import numpy as np
import os
```

```python
root_path = os.getcwd()
dir_path = os.path.join(root_path, "datasets")
filename = "board_games.csv"
file_path = os.path.join(dir_path, filename)
```

```python
df = pd.read_csv(file_path)
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df_ = df.loc[:, ['game_id', 'min_players', 'year_published', 'users_rated']]
df_
```

```python
multiplayer_games = df_[df_['min_players'] >= 2]
multiplayer_games = multiplayer_games.filter(like = '_', axis = 1)
multiplayer_games
```

```python
multiplayer_games.boxplot(column = 'year_published')
```

```python
multiplayer_games.sort_values(by = ["year_published", "game_id"], ascending=False)
```

```python
multiplayer_games.rename(columns = {'game_id': 'id', 'year_published': 'release_year'}, inplace = True)
multiplayer_games
```
